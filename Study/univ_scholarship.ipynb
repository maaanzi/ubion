```python
import csv

file_name = 'csv/한국장학재단_대학별 장학금 수혜 현황_20230831.csv'
fp = open(file_name, 'r', encoding = 'cp949')
```

```python
data = []
rd = csv.reader(fp)
for line in rd :
    data.append(line)

for each_line in data[:5] : # 0,1,2,3,4
    print(each_line)
```

```python
col = data[0]
cnt = 0

search = input("검색할 학교명 : ")
for line in data [1: ] :
    if search in line[col.index('학교명')] : # 한줄씩 확인, in : 포함되어 있는지 확인
        cnt += 1
        print(str(cnt) + ".")
        for i, name in enumerate(col) :
            print(f"{name} : {line[i]}")
        print('=' * 40)

print(f"총 {cnt}건이 확인되었습니다.")
```
