```python
import requests

url = 'https://eiec.kdi.re.kr/policy/materialList.do'

response = requests.get(url, verify=False) # 신뢰할 수 없는 인증서 : SSLError: HTTPSConnectionPool(host='eiec.kdi.re.kr',
print(response.text)
```

```python
from bs4 import BeautifulSoup

html = response.text

soup = BeautifulSoup(html, 'html.parser')

list_title = soup.find_all('div',{'class' : 'list_txt'})
# list_title.find('span').decompose()

data_list = []

for title in list_title :
    # print(title.get_text(strip=True))
    # data = title.get_text(strip=True)
    data = title.find('p').get_text()
    data_list.append(data)
print(data_list)
```
