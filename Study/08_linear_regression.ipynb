```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
model = LinearRegression()

X = [[1],[2],[3],[4],[5]]
y = [[3],[5],[9],[13],[15]]
model.fit(X,y)
```

```python
# y = 3.2x - 0.6
print('coef : ', model.coef_) # 기울기
print('intercept : ', model.intercept_) # 절편
```

```python
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(X,y)
plt.plot(X,model.coef_*X + model.intercept_)
```

```python
x = [[6]]
y_predict = model.predict(x)
print('x : {} -> y_predict : {}'.format(x,y_predict))
```

# 붓꽃 데이터 불러오기

```python
from sklearn.datasets import load_iris

iris = load_iris()

# sepal(꽃받침) length, sepal width, petal(꽃잎) length, petal width
X = iris.data
print(f'데이터 수 : {len(X)}\n데이터 >>\n {X}')
```

```python
y = iris.target # 레이블(결정값) 데이터를 가져온다
print(f'데이터 수 : {len(y)}\n데이터 >>\n {y}')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2) # 0.2 : 20%는 테스트데이터로 빼겠다, 80%는 학습데이터
print(f'학습 결과 값 >>\n데이터 수 : {len(y_train)}\n {y_train}\n학습 데이터 >> \n{X_train}')
```
