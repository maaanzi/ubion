```python
# 웹페이지 요청을 위한 패키지
import requests
```

```python
url = 'https://news.naver.com/'

# 뉴스 페이지 요청, 저장
response = requests.get(url)
print(response.text) # 내용 확인
```

```python
# 웹페이지 파싱을 위해
from bs4 import BeautifulSoup

# response로 부터 html 가져오기
html = response.text

soup = BeautifulSoup(html, 'html.parser')

# html에서 원하는 태그 찾기
hdline_title = soup.find_all('div',{'class' : 'cjs_t'})

for title in hdline_title :
    print(title.get_text(strip=True)) # 결과 출력 (뉴스 제목)
```

```python
# 텍스트
file = open('csv/TheLittlePrince.txt', 'r', encoding='CP949')
text = file.read()
print(text)
```

```python
word_list = text.split()

count_list = []
for word in word_list :
    count = word_list.count(word)
    count_list.append([word, count])

sorted_list = sorted(count_list, key = lambda x:x[1], reverse = True) # reverse = True : 내림차순 정렬

word_dict = {}
for word, frequency in sorted_list :
    if word not in word_dict :
        word_dict[word] = frequency
        print(word, frequency)
```

```python
from pptx import Presentation

contents = Presentation('csv/database.pptx')
text_collection = []
for slide in contents.slides :
    for shape in slide.shapes: # 슬라이드 하나 당 있는 오브젝트들(사진, 글 등등)
        if not shape.has_text_frame : # 텍스트인지 확인
            continue
        for paragraph in shape.text_frame.paragraphs :
            text_collection.append(paragraph.text)
print(text_collection)
```

```python
outfp = open('text_data.txt','w', encoding='utf-8')
outfp.writelines(text_collection)
outfp.close()

infp = open('text_data.txt', 'r',encoding='utf-8')
data = infp.read()
print(data)
```
