```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.korea.kr/news/top50List.do'
response = requests.get(url)

soup = BeautifulSoup(response.text)
top_news = []

# text(클래스) -> span -> strong 태그 선택
strong = soup.select('span.text > strong')
for n in strong :
    top_news.append(n.get_text().strip())

print(top_news)
```

```python
from datetime import datetime
year = datetime.today().year
month = datetime.today().month
day = datetime.today().day

print(f'<<< {year}년 {month}월 {day}일', end=' ')
print('대한민국 정책 브리핑 TOP 50 뉴스 >>>')

n = 1
for i in top_news[:50] :
    print("%d: "%n+i) # 앞 번호 매기기
    n = n + 1
```

```python
file_name = '../csv/먼 후일.txt'
fp = open(file_name, 'r',encoding='utf-8')

data = fp.readlines() # 리스트로 저장
fp.close()

print(data)
```

```python
for line in data :
    print(line, end='')
```

```python
fp2 = open(file_name.split(".")[0]+"_작업본.txt",'w')

cnt = 0
where = []

for i, line in enumerate(data):
    if '잊었노라' in line :
        cnt += 1
        where.append(i)
        fp2.write(str(i) + ' : ' + line)

fp2.close()

print(f"'잊었노라'는 {where}줄에 총 {cnt}번 존재합니다.")
```
