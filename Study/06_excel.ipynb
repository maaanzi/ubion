```python
import openpyxl

excel_file = openpyxl.load_workbook('csv/netflix_data.xlsx')
```

```python
selected_sheet = excel_file.active
print(selected_sheet) # 시트 확인
```

```python
excel_file.sheetnames
```

```python
for sheet in excel_file :
    print(sheet.title)
```

```python
data_col = selected_sheet['L'] # L열 의미
for each in data_col :
    print(each.value)
```

```python
data_row = selected_sheet[1]
for each in data_row :
    print(each.value)
```

```python
sel_data = selected_sheet['C2' : 'L2'] # c2 ~ l2
for data in sel_data :
    for each in data :
        print(each.value)
```

```python
sel_data = selected_sheet['C':'E']
for data in sel_data :
    for each in data :
        print(each.value, end = ',') # 줄바꿈 없이 ,로 구분
    print("\n\n >>> The Next Column")
```

```python
# iter_cols() : 반복
for data in selected_sheet.iter_cols(min_col=1, max_row=4, max_col=5):
    for each in data :
        print(each.value, end=',')
    print("\n\n >>> The Next Column")
```

```python
sel_data = selected_sheet[1:3]
for data in sel_data :
    for each in data :
        print(each.value, end=',')
    print("\n\n >>> The Next Column")
```

```python
for data in selected_sheet.iter_rows(min_row=1, min_col=3, max_col=5, max_row=5) :
    for each in data :
        print(each.value, end = ',')
    print("\n\n >>> The Next Row")
```
