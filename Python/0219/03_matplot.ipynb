```python
# 라인그래프 : plot({1차원 데이터 형태})
# plot 1차원 데이터 인자의 개수가 1개라면 y축 데이터로 대입, x축은 0부터 1씩 증가 (2개라면 x, y)
data = [5,3,7,1]

plt.plot(data)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
```

```python
drinks = pd.read_csv('../../csv/drinks.csv')
```

```python
drinks.info()
```

```python
# 상관계수 출력 : corr()
corr = drinks.iloc[:, 1:5].corr(method='pearson')
corr
```

```python
# 히트맵 출력
cols_view = ['beer','spirit','wine','alcohol']

hm = sns.heatmap(
    corr.values,
    cbar = True, # 사이드바 유무
    annot = True, # 히트맵 숫자 표시 유무
    square = True, # 정사각형 표시 유무
    fmt = '.3f', # 숫자의 소수점 자리수
    annot_kws = {'size' : 10},
    xticklabels = cols_view,
    yticklabels = cols_view,
    cmap = 'Greens'
)
```

```python
# 분포도 시각화
sns.pairplot(
    drinks.loc[:, 'beer_servings' : 'total_litres_of_pure_alcohol'],
    height=2.5, # 표의 크기
    corner=True, # 대칭되는것 제외?
    kind= 'reg',
    plot_kws= {
        'line_kws' : {
            'color' : 'red'
        }
    }
)

plt.tight_layout()
plt.show()
```
