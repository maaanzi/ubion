```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

svc_clf = SVC()

parameters = {
    "C" : [1,10,100],
    "gamma" : [0.1,0.01,0.001]
}

# cv : 교차검증 반복 횟수
gscv_scv = GridSearchCV(estimator=svc_clf, param_grid=parameters, scoring='accuracy', cv=3, refit=True, n_jobs=1, verbose=1)
gscv_scv.fit(X_train, y_train)
```

```python
gscv_scv.best_params_
```

```python
gscv_scv.best_score_
```

```python
from sklearn.metrics import accuracy_score

y_pred = gscv_scv.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
pd.DataFrame(data=cm, columns=['Real P:1', 'Real N:0'], index=['Predict P:1', 'Predict N:0'])
```
